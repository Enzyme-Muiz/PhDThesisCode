```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../../data/other_data_for_20_cll/concattransformedPlayR.csv", index_col=0)
data.columns
```

```python
correlation_matrix = data.corr(method='spearman')
```

```python
# Create a heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix, 
            #annot=True, 
            cmap='coolwarm', 
            linewidths=0.5)
plt.title('Spearman Correlation Heatmap of the Markers')
plt.show()
```

```python
sorted_df = correlation_matrix.stack().reset_index().rename(columns={0: 'Value'})
sorted_df = sorted_df.sort_values(by='Value', ascending=False)

# Get the top 10 distinct highest values
top_10_distinct_values = sorted_df['Value'].unique()[:10]

print(top_10_distinct_values)
```
